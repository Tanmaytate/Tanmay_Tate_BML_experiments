```python
# ðŸ“Œ Step 1: Import Libraries
import sqlite3
import pandas as pd

# ðŸ“Œ Step 2: Connect to SQLite Database
# This creates a new database file named 'lab_data.db' if it doesn't exist
conn = sqlite3.connect("lab_data.db")
cursor = conn.cursor()

# ðŸ“Œ Step 3: Create a Sample Table (Optional)
# You can skip this step if your database already has tables
cursor.execute("""
CREATE TABLE IF NOT EXISTS students (
    id INTEGER PRIMARY KEY,
    name TEXT,
    marks INTEGER
)
""")

# ðŸ“Œ Step 4: Insert Sample Data (Optional)
# This is just for demonstrationâ€”skip if your table already has data
cursor.execute("INSERT INTO students (name, marks) VALUES ('Alice', 85)")
cursor.execute("INSERT INTO students (name, marks) VALUES ('Bob', 78)")
cursor.execute("INSERT INTO students (name, marks) VALUES ('Charlie', 92)")
conn.commit()

# ðŸ“Œ Step 5: Extract Data Using SQL Query
query = "SELECT * FROM students"
df = pd.read_sql_query(query, conn)

# ðŸ“Œ Step 6: Display the Extracted Data
print("ðŸ“Š Extracted Data from Database:")
print(df)

# ðŸ“Œ Step 7: Close the Connection
conn.close()
```
